```python
import requests
from bs4 import BeautifulSoup
import xlsxwriter

# Read URLs from a text file
with open('links.txt', 'r') as file:
    urls = file.read().splitlines()

# Create an Excel workbook and worksheet using xlsxwriter
excel_file_path = 'poetry8.xlsx'
workbook = xlsxwriter.Workbook(excel_file_path)
worksheet = workbook.add_worksheet()

# Initialize row number
row_num = 0

# Loop through each URL
for url in urls:
    # Fetch HTML content
    response = requests.get(url)
    if response.status_code == 200:
        html_content = response.text
        print(f'Successfully fetched content from: {url}')
    else:
        print(f'Failed to fetch HTML content for URL: {url}')
        continue

    # Parse HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all elements with the CSS class 'has-text-align-right'
    elements_with_right_class = soup.find_all(class_='has-text-align-center')
    
    # Find elements with the CSS class 'entry-title'
    elements_with_title_class = soup.find_all(class_='entry-title')

    # Extract data from each element and concatenate it into a single string
    text_right_class = '\n'.join([element.get_text() for element in elements_with_right_class])
    text_title_class = '\n'.join([element.get_text() for element in elements_with_title_class])

    # Write the concatenated text to the worksheet in a single cell
    worksheet.write(row_num, 0, text_right_class)
    worksheet.write(row_num, 1, text_title_class)
    row_num += 1

# Close the workbook
workbook.close()

print(f'Scraped content with "has-text-align-right" class saved to {excel_file_path}')
```
